```python
import csv

file_menu = 'seblak.csv'
file_pesanan = 'pesanan.csv'

def baca_menu():
    # Membaca item-menu dari file CSV
    with open(file_menu, 'r') as file:
        pembaca = csv.reader(file)
        menu_items = list(pembaca)
    return menu_items

def tulis_pesanan(pesanan):
    # Menulis pesanan ke file CSV
    with open(file_pesanan, 'a', newline='') as file:
        penulis = csv.writer(file)
        penulis.writerow(pesanan)

def tampilkan_menu():
    # Implementasikan fungsionalitas tampilan menu restoran di sini
    print("Selamat datang di Restoran Kami!")
    print("1. Lihat Menu")
    print("2. Pesan Makanan")
    print("3. Lihat Riwayat Pesanan")
    print("4. Keluar")

def lihat_menu():
    # Implementasikan fungsionalitas tampilan menu di sini
    menu_items = baca_menu()
    print("Menu:")
    for indeks, item in enumerate(menu_items, start=1):
        print(f"{indeks}. {item[0]} - Rp {item[1]}")

def pesan_makanan():
    # Implementasikan fungsionalitas pemesanan makanan di sini
    nomor_item = int(input("Masukkan nomor item untuk dipesan: "))
    jumlah = int(input("Masukkan jumlah: "))
    
    # Menambahkan pesanan ke file CSV pesanan
    pesanan = [nomor_item, jumlah]
    tulis_pesanan(pesanan)

    print(f"Pesanan ditempatkan: {jumlah}x {menu_items[nomor_item-1][0]}")

def lihat_riwayat_pesanan():
    # Membaca riwayat pesanan dari file CSV
    with open(file_pesanan, 'r') as file:
        pembaca = csv.reader(file)
        pesanan = list(pembaca)

    print("Riwayat Pesanan:")
    for pesan in pesanan:
        nomor_item, jumlah = pesan
        print(f"Item: {menu_items[int(nomor_item)-1][0]}, Jumlah: {jumlah}")

if __name__ == "__main__":
    running = True
    menu_items = baca_menu()  # Membaca menu sekali saat program dimulai

    while running:
        tampilkan_menu()
        input_user = input("Pilih opsi (1-4): ")

        if input_user == "1":
            lihat_menu()

        elif input_user == "2":
            pesan_makanan()

        elif input_user == "3":
            lihat_riwayat_pesanan()

        elif input_user == "4":
            running = False
            print("Terima kasih telah menggunakan sistem pemesanan restoran kami!")

        else:
            print("Opsi tidak valid. Silakan pilih opsi yang valid (1-4).")
```
